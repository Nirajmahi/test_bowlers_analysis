```python
import pandas as pd
```

```python
hw1= pd.read_csv("Untitled spreadsheet - Sheet1.csv")
display(hw1.head())
```

#####  Removing BBI & BBM columns

```python
hw1= hw1.drop('BBI',axis=1)
hw1= hw1.drop('BBM',axis=1)
display(hw1.head())
```

##### Extracting the country information from player column

```python
hw= hw1['Player'].str.split("(",expand=True)
hw1=pd.concat([hw1,hw],axis=1)
hw1=hw1.drop('Player',axis=1)
hw1=hw1.rename(columns={0:'Player',
                       1:'Country'})
hw1['Country']=hw1['Country'].str.replace(")", "")
new_seq= ['Player', 'Span', 'Mat', 'Inns', 'Balls', 'Runs', 'Wkts',
                    'Ave', 'Econ', 'SR', '5','10' ,'Country']
hw1= hw1[new_seq]
display(hw1.head())
```

##### Creating a column to check whether  the player played ICC  or not

```python
def is_Icc(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
hw1['is_ICC'] = hw1['Country'].apply(is_Icc)
display(hw1.head(10))
```

#### No. of players played for ICC

```python
(hw1.is_ICC=='Yes').sum()
```

#####  5 playes played for ICC

##### Calculating the span of a player

```python
span = hw1['Span'].str.split("-", expand=True)
hw1 = pd.concat([hw1, span], axis=1)
hw1 = hw1.rename(columns={0: "Begin",
                        1: "End"})
hw1 = hw1.drop("Span", axis=1)

display(hw1.head())
```

#####  Beginning & Ending years are now in Begin and End Columns

```python
hw1['Begin'] = hw1['Begin'].astype('int') 
hw1['End'] = hw1['End'].astype('int')
```

 #### calculating  the span of a player under column  total_years

```python
hw1['total_years'] = hw1['End'] - hw1['Begin']

hw1 = hw1.drop(['Begin', "End"], axis=1)

display(hw1.head())
```

#### Player(s) played for the longest period
